# Kenya Crop Yield Analysis

## Project Overview

- This project analyzes agricultural data from different crops grown in Kenya to identify patterns in yield, revenue, and profitability.

- The analysis explores how crop type, weather conditions, farm characteristics, and production factors influence agricultural outcomes.

## Objectives

- Analyze production trends across different crops in Kenya
- Compare crop yields and profitability
- Identify factors affecting farm revenue and profit
- Generate insights that can support agricultural decision-making

## Key Findings

- Cassava recorded the highest yield among the analyzed crops.
- Weather conditions and farming factors influence production outcomes.
- Profitability varies significantly across crop types.

```python
# importing necessary library
import pandas as pd
#reading the dataset into python
df=pd.read_csv('Kenya_Crops_Dataset.csv')
# viewing the first 10 r0ws
df.head(10)
```

```python
# viewing the shape of the dataset
df.shape
```

```python
# checking data types and missing values
df.info()
```

```python
df.head()
# viewing first five rows to be able to change the data types correctly
```

```python
# checking missings
df.isnull().sum()
```

```python
# checking duplicates
df.duplicated().sum()
```

```python
# finding the "Error" values
(df == "Error").sum()
```

```python
# replacing all "Error" with NaN
import numpy as np

df.replace("Error", np.nan, inplace=True)
```

```python
# finding the "Error" values
(df == "Error").sum()
```

```python
df.info()
```

```python
numeric_columns = [
    'Planted Area (Acres)',
    'Yield (Kg)',
    'Market Price (KES/Kg)',
    'Cost of Production (KES)'
]

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric)
```

```python
# filling missing values with median for numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
df[numeric_cols].isnull().sum()
```

```python
df['Planting Date'] = pd.to_datetime(df['Planting Date'], errors='coerce')
df['Harvest Date'] = pd.to_datetime(df['Harvest Date'], errors='coerce')
```

```python
# working with categorical columns, clleaning
df.select_dtypes(include='object').isnull().sum()
```

```python
# replacing missings in categorical columns with mode
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# checking if there are any missings
df.isnull().sum()
```

```python
# checking rows with missing dates
df[['Planting Date', 'Harvest Date']].isnull().sum()
```

```python
# dropping rows with missing dates
df = df.dropna(subset=['Planting Date', 'Harvest Date'])
```

```python
# which crop type has the highest yield
import matplotlib.pyplot as plt

avg_yield = df.groupby('Crop Type')['Yield (Kg)'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
bars = plt.bar(avg_yield.index, avg_yield.values)

plt.title('Average Yield by Crop Type', fontsize=16)
plt.xlabel('Crop Type')
plt.ylabel('Average Yield (Kg)')
plt.xticks(rotation=45)

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 20,
        f'{height:.0f}',
        ha='center',
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

## Insight

- The analysis shows that cassava recorded the highest average yield(2,899 kg) among all crop types, followed closely by rice(2,889 kg) and sorghum.(2,837 kg)
- Wheat produced the lowest average yield.(2,005 kg)
- These findings suggest that cassava may be the most productive crop under the farming conditions represented in this dataset.
- However, yield alone does not determine profitability, as production costs and market prices also influence farmers' returns.

```python
# saving my chart
plt.savefig("average_yield_by_crop_type.png", dpi=300, bbox_inches="tight")
```

```python
# which county generates the highest average profit
avg_profit = df.groupby('County')['Profit (KES)'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
bars = plt.bar(avg_profit.index, avg_profit.values)

plt.title("Average Profit by County")
plt.xlabel("County")
plt.ylabel("Average Profit (KES)")
plt.xticks(rotation=45)

for bar in bars:
    plt.text(
        bar.get_x()+bar.get_width()/2,
        bar.get_height(),
        f'{bar.get_height():,.0f}',
        ha='center',
        fontsize=8
    )

plt.tight_layout()
plt.show()
```

```python
plt.savefig("average_profit_by_county.png", dpi=300, bbox_inches="tight")
```

### Insight

- Nyeri recorded the highest average profit (KES 3,446,646), followed by Kiambu (KES 3,051,997).
- Machakos generated the lowest average profit among the counties analyzed.

- The variation in profitability suggests that geographical factors such as climate, soil quality, farming practices, market access, and crop selection may influence farm income.

- Counties with consistently higher profits could provide useful benchmarks for identifying successful agricultural practices that could be adopted elsewhere.

```python
# which season produces the highest yield
season_yield = df.groupby('Season')['Yield (Kg)'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
bars = plt.bar(season_yield.index, season_yield.values)

plt.title("Average Yield by Season")
plt.xlabel("Season")
plt.ylabel("Average Yield (Kg)")

for bar in bars:
    plt.text(
        bar.get_x()+bar.get_width()/2,
        bar.get_height()+15,
        f'{bar.get_height():,.0f}',
        ha='center'
    )

plt.tight_layout()
plt.show()
```

```python
plt.savefig("average_yield_by_season.png", dpi=300, bbox_inches="tight")
```

### Insight

- The **Short Rains** season recorded the highest average crop yield (**2,731 kg**), followed by the **Dry Season** (**2,606 kg**) and the **Long Rains** (**2,562 kg**).

- Although the differences in average yield are relatively small, the results suggest that crops grown during the Short Rains season may benefit from more favorable growing conditions.
- Factors such as rainfall distribution, temperature, crop selection, and farming practices could contribute to this higher productivity.

- These findings indicate that seasonality should be considered when planning crop production to maximize yields.

```python
# which irrigation method is associated with the highest profit?
irrigation_profit = df.groupby('Irrigation Method')['Profit (KES)'].mean().sort_values(ascending=False)

plt.figure(figsize=(9,5))
bars = plt.bar(irrigation_profit.index, irrigation_profit.values)

plt.title("Average Profit by Irrigation Method")
plt.xlabel("Irrigation Method")
plt.ylabel("Average Profit (KES)")

for bar in bars:
    plt.text(
        bar.get_x()+bar.get_width()/2,
        bar.get_height(),
        f'{bar.get_height():,.0f}',
        ha='center',
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

```python
plt.savefig("average_profit_by_irrigation_method.png", dpi=300, bbox_inches="tight")
```

### Insight

- The analysis shows that farms using **Sprinkler irrigation** recorded the highest average profit (KES 2,647,348), followed by **Flood irrigation** (KES 2,567,034). **Drip irrigation** generated the lowest average profit (KES 2,505,246).

- Although sprinkler irrigation appears to be associated with higher profitability in this dataset, the differences between the three irrigation methods are relatively small. This suggests that irrigation method alone may not determine profitability.
- Other factors such as crop type, production costs, farm size, soil conditions, and market prices are also likely to influence farm income.

```python
# does planting larger area lead to higher yields?
plt.figure(figsize=(8,6))

plt.scatter(
    df['Planted Area (Acres)'],
    df['Yield (Kg)']
)

plt.title("Planted Area vs Yield")
plt.xlabel("Planted Area (Acres)")
plt.ylabel("Yield (Kg)")

plt.show()
```

### Insight

- The scatter plot shows no strong relationship between planted area and crop yield. Farms with both small and large planted areas recorded a wide range of yields.

- This suggests that simply increasing the amount of land planted does not guarantee higher crop yields. Other factors such as crop type, soil quality, irrigation method, weather conditions, and farming practices are likely to have a greater influence on productivity.

- Further statistical analysis, such as calculating the correlation coefficient, would help quantify the strength of the relationship between planted area and yield.
